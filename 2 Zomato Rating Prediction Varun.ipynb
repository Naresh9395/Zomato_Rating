# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.dtypes
```

```python
def Encode(df):
    for column in df.columns[~df.columns.isin(['name','rate', 'votes','location', 'rest_type','cuisines','cost'])]:
        df[column] = le.fit_transform(df[column])
    return df

df_encoded = Encode(df.copy())
```

# Model: Random Forest Regressor
